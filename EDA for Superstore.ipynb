```python
profitcategory=data[profit].groupby(['Category']).sum().sort_values(by='Profit',ascending=False)
profitcategory
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=data,x='Profit',y='Discount')
plt.title('RelationShip between Profit and Discount')
plt.show()
```
